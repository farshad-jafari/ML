```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
x_5 = np.random.rand(5, 1) * 20
for i in range (0,len(x_5)):
    for j in range (i+1, len(x_5)):
        if (float(x_5[i]) > float(x_5 [j])):
            temp = float(x_5[i])
            (x_5[i])  = float(x_5[j])
            (x_5[j]) = temp
y_5 = 2.358 * x_5 - 3.121
y_5 = y_5 + np.random.normal(scale=3, size=y_5.shape)
dataset_5 = np.hstack((x_5, y_5))
MSE = np.zeros ((4 , 11))
```

```python
x_10 = np.random.rand(10, 1) * 20
for i in range (0,len(x_10)):
    for j in range (i+1, len(x_10)):
        if (float(x_10[i]) > float(x_10 [j])):
            temp = float(x_10[i])
            (x_10[i])  = float(x_10[j])
            (x_10[j]) = temp
y_10 = 2.358 * x_10 - 3.121
y_10 = y_10 + np.random.normal(scale=3, size=y_10.shape)
dataset_10 = np.hstack((x_10, y_10))
```

```python
x_25 = np.random.rand(25, 1) * 20
for i in range (0,len(x_25)):
    for j in range (i+1, len(x_25)):
        if (float(x_25[i]) > float(x_25 [j])):
            temp = float(x_25[i])
            (x_25[i])  = float(x_25[j])
            (x_25[j]) = temp
y_25 = 2.358 * x_25 - 3.121
y_25 = y_25 + np.random.normal(scale=3, size=y_25.shape)
dataset_25 = np.hstack((x_25, y_25))
```

```python
x_100 = np.random.rand(100, 1) * 20
for i in range (0,len(x_100)):
    for j in range (i+1, len(x_100)):
        if (float(x_100[i]) > float(x_100 [j])):
            temp = float(x_100[i])
            (x_100[i])  = float(x_100[j])
            (x_100[j]) = temp
y_100 = 2.358 * x_100 - 3.121
y_100 = y_100 + np.random.normal(scale=3, size=y_100.shape)
dataset_100 = np.hstack((x_100, y_100))
```

```python
# Fitting Linear Regression to the dataset 
from sklearn.linear_model import LinearRegression 
lin_5 = LinearRegression() 
lin_5.fit(x_5, y_5) 
predicted_list  = lin_5.predict(x_5)
MSE [0][0] = mean_squared_error(y_5, predicted_list)
print("MSE(mean squared error)", mean_squared_error(y_5, predicted_list))
lin_10 = LinearRegression() 
lin_10.fit(x_10, y_10)
predicted_list  = lin_10.predict(x_10)
MSE [1][0] = mean_squared_error(y_10, predicted_list)
print("MSE(mean squared error)", mean_squared_error(y_10, predicted_list))
lin_25 = LinearRegression() 
lin_25.fit(x_25, y_25) 
predicted_list  = lin_25.predict(x_25)
MSE [2][0] = mean_squared_error(y_25, predicted_list)
print("MSE(mean squared error)", mean_squared_error(y_25, predicted_list))
lin_100 = LinearRegression() 
lin_100.fit(x_100, y_100) 
predicted_list  = lin_100.predict(x_100)
MSE [3][0] = mean_squared_error(y_100, predicted_list)
print("MSE(mean squared error)", mean_squared_error(y_100, predicted_list))
```

```python
# Visualising the Linear Regression results

plt.scatter(x_5, y_5, color = 'blue') 
  
plt.plot(x_5, lin_5.predict(x_5), color = 'red') 
plt.title('Linear Regression, x_5') 
plt.xlabel('x_5_data') 
plt.ylabel('y_5_Linear_Regression') 
  
plt.show() 

plt.scatter(x_10, y_10, color = 'blue') 
  
plt.plot(x_10, lin_10.predict(x_10), color = 'red') 
plt.title('Linear Regression, x_10') 
plt.xlabel('x_10_data') 
plt.ylabel('y_10_Linear_Regression') 
  
plt.show() 

plt.scatter(x_25, y_25, color = 'blue') 
  
plt.plot(x_25, lin_25.predict(x_25), color = 'red') 
plt.title('Linear Regression, x_25') 
plt.xlabel('x_25_data') 
plt.ylabel('y_25_Linear_Regression') 
  
plt.show() 

plt.scatter(x_100, y_100, color = 'blue') 
  
plt.plot(x_100, lin_100.predict(x_100), color = 'red') 
plt.title('Linear Regression, x_100') 
plt.xlabel('x_100_data') 
plt.ylabel('y_100_Linear_Regression') 
  
plt.show() 
```
